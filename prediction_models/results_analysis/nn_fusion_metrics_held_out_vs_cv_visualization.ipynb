```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load model metrics
held_out_metrics = pd.read_csv('../model_results/nn_fusion_clf_results/20260707_110649/intermediate_fusion_nn_results.csv')
cv_metrics = pd.read_csv('../model_results/nn_fusion_clf_results_cross_val/20260707_111545/intermediate_fusion_nn_results_aggregated.csv')
```

```python
metrics = ['accuracy', 'f1', 'precision', 'recall', 'auc']
embedders = held_out_metrics['embedding'].tolist()

held_out_metrics = held_out_metrics.set_index('embedding')
cv_metrics = cv_metrics.set_index('embedding')
```

```python
x = np.arange(len(embedders))
width = 0.35
 
for metric in metrics:
    fig, ax = plt.subplots(figsize=(10, 5))
 
    held_vals = held_out_metrics.loc[embedders, f'test_{metric}'].values
    agg_vals = cv_metrics.loc[embedders, f'test_{metric}_mean'].values
    agg_std = cv_metrics.loc[embedders, f'test_{metric}_std'].values
 
    ax.bar(x - width/2, held_vals, width, label='Held-out set (single split)', color='#4C72B0')
    ax.bar(x + width/2, agg_vals, width, yerr=agg_std, capsize=4,
           label='10-fold cross-val (mean ± std)', color='#DD8452')
 
    ax.set_ylabel(metric.upper())
    ax.set_title(f'Intermediate Fusion NN: Test {metric.upper()} by Embedder', fontsize=13)
    ax.set_ylim(0.75, 1.0)
    ax.grid(axis='y', alpha=0.3)
    ax.set_xticks(x)
    ax.set_xticklabels(embedders, rotation=30, ha='right')
    ax.legend(loc='lower right')
 
    fig.tight_layout()
    fig.savefig(f'./visualizations/held_out_vs_cv_{metric}.png', dpi=150, bbox_inches='tight')
    plt.close(fig)
    print(f"Saved {metric} plot.")
```
